```python
import os

input_dir = r"D:\4th year project\Data\Sample set 4 - start of 2nd semester\0h(2)"
output_dir = r"D:\4th year project\Data\Sample set 4 - start of 2nd semester\0h(2)\reformatted data"

# Create output folder if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through all files in input folder
for filename in os.listdir(input_dir):
    if filename.endswith(".txt"):

        input_file = os.path.join(input_dir, filename)
        output_file = os.path.join(output_dir, filename)

        with open(input_file, "r") as infile, open(output_file, "w") as outfile:
            start_reading = False

            for line in infile:
                line = line.strip()

                if line.startswith(">>>>>Begin Spectral Data<<<<<"):
                    start_reading = True
                    continue

                if start_reading and line:
                    parts = line.split()

                    if len(parts) == 2:
                        wavelength = float(parts[0])
                        intensity = float(parts[1])

                        intensity = round(intensity, 2)

                        outfile.write(f"{wavelength:.3f}\t{intensity:.2f}\n")

print("All files processed.")
```
